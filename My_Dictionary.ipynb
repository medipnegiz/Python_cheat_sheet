### **SET (KÜME) FORMATI KOD KULLANIMLARI**

```python
a = {'orange', 'blau', 'red', 'blau', 'rose'}                     #   "{}" veya "set" sarmalina alarak olusturulabilir. Sirasiz bir sekilde yazdirir.
flower = 'rose', 'orkhide', 'rose', 'violet', 'red'                                                        
b = set(flower)                                                   #   "set" küme icindeki elemanlarin hepsinden birer tane gösterir. tekrarlama olmaz
bos_kume = set()                                                  #   bos küme icin "set" sarmalina alinmali. sadece "{}" olursa dict olarak algilar
c = set('philadelphia')                                           #   icinde tek öge varsa hepsini ayirir, tekrarsiz ve sirasiz yazar. (icine tek deger alir)
d =set('dlia')
e = {'dolphin'}                                                   #   icindeki ögeleri tek olarak algilar ve yazar.
print('1:', a.difference(b))                                      #   Iki kümeyi karsilastirir sadece "a" da farkli olanlari yazar (Fark)
print('2:', b.difference(a))                                      #   Iki kümeyi karsilastirir sadece "b" de farkli olanlari yazar (Fark)
print('3:', a - b)                                                #   Iki kümeyi karsilastirir sadece farkli olanlari yazar

print('4:', a.union(b))                                           #   Iki kümenin birlesimini verir. Tekrar olanlari yazmaz (Birlesim)
print('5:', a | b )                                               #   Iki kümenin birlesimini verir. Tekrar olanlari yazmaz (Birlesim)

print('6:', a.intersection(b))                                    #   Sadece iki kümede de olanlari yazar (Kesisim)
print('7:', a & b)                                                #   Sadece iki kümede de olanlari yazar (Kesisim)
bos_kume = frozenset(bos_kume)                                    #   "frozenset" bu küme formatinin icine ekleme yapilamaz
a.remove('orange')                                                #   Küme icerisinden istenilen degei siler. Deger kümede yoksa hata verir.
new = a.copy()                                                    #   A kümesindeki bütün elemanlari yeni kümeye kopyalar
a.add('yellow')                                                   #   Küme icerisine ekleme yapar
print('8:', a)
print('9:', 'red' in b)                                           #   Kümenin icinde istedigimiz kelime var mi diye bakar
print('10:', 'purple' not in b)                                   #   Kümenin icinde istedigimiz kelime yok mu diye dogrulama yapilir
print('11:', c)
print('12:', e)
print('13: silindi..:', c.pop())                                  #   küme ile kullanimda deger istemiyor. Rastgele silme yapiyor
c.discard('p')                                                    #   küme icinde istenilen degeri siler. Deger bulamazsa hicbir sey yapmaz.
print('14:', d.issubset(c))                                       #   d kümesi c kümesinin alt kümesi mi diye sorgular ve True-False verir
print('15:', c.issuperset(d))                                     #   c kümesi d kümesini kapsar mi diye sorgular ve True-False verir
```

### **IF ve INPUT KOMUT KULLANIMLARI**



```python
score = int(input('Puaninizi girin..: '))       #   kullanici tarafindan girdi yapilmasini saglar. aksi belirtilmedikce "string" ifade tanimlar
if score >= 90:                                 #   her zaman bir sarta baglidir ve satir sonunda " : " isareti gelir. Belirtilmezse ilk "True" mu diye bakar. 
    if score >= 95:                             #   ic ice if komutu verilerek dallanma islemi yapilabilir
        derece = 'A+'                           #   if tanim ve yazdirmada her zaman bosluk birakilir ve icten yazmaya baslanir
    else:
        derece = 'A'
elif score >= 80:                               #   bu komutun "eger öyle degil böyleyse" anlami vardir.
        derece = 'B'
elif score > 70 or score == 70:                 #   "or" komutu ile ikisinden birisi dogru ise anlami verilebilir
        derece = 'C' 
elif score >= 50 and score <= 60:               #   "and" komutu ile iki degerde dogru ise islemi yap komutu verdik                  
        pass                                    #   "pass" komutu herhangi bir islemi pas gecmek icin kullanilir.
elif score == 40:
    raise  Exception('Bu nasil not..:(')        #   "raise" istenmeyen bir giriste hata verdirmek icin kullanilir (TypeError, ValueError)
else:                                           #   bu komutun "bunlarin hicbiri degilse" anlami vardir.
    derece = 'Gecemediniz'
print('Dereceniz..: ', derece)

sayi = []
for i in range(10):
    if i % 2 == 0:                            #   Cift sayilari listeye ekler, tek sayilarin karesini listeye ekler.
        sayi.append(i)
    else:
        sayi.append(i**2)
print('1:', sayi)

yeni = [i if i % 2 == 0 else i**2 for i in range(10)]           #   for, if ve else komutlarini tek satirda yazdik. Siralama önemli.
print('2:', yeni)

sayi = 50
print('3:','Sayi 50' if sayi == 50 else 'Sayi belirsiz')        #   Tek satirda if ve else komutu kullanimi
```

### **FOR DÖNGÜ KOMUT KULLANIMI**

```python
#  For döngüsü genellikle liste, tanımlama grubu, küme ve sözlük gibi çeşitli koleksiyon türlerinin öğelerini yinelemek için kullanılır.

season = ['spring', 'summer', 'autumn', 'winter']
for a in season:                                    #   liste icinde gezerek hepsini teker teker yazdirir
    print('1:', a)                 

dongu = list(range(1,8))                            #   " range " istenilen rakama kadar dongu olusturur. (baslangic, bitis, kacar adim) olarak verilebilir
para = 1000
for i in dongu:                                     #   liste icinde döngü yapar yada sayi üretir bir kosula bagli degildir. burada icindekiler kismi tanimlaniyor
    para += para * (0.07)
    i += 1                                          #   burada "i" tanimlamaya gerek yok kendisini otomatik olarak basliyor
print('2:', int(para), '$')                         #   print bosluklari önemli ona göre tüm döngüyü yazar ya da sonucu yazar
print('3:', dongu)

who = ['Ich bin ', 'Du bist ']
mood = ['froh', 'angry', 'müde']
for i in who:                                       #   ic ice for döngüsü olusturulabilir. 
    for c in mood:                                  #   dis döngü baslar ilk degerini alir ic döngüyü bitirir sonra dis döngü siradaki degeri alir
        print('4:', i, c)

text = ['eins', 'zwei', 'drei']
num = [1, 2, 3, 4]
for x, y in zip(text, num):                         #   "zip" komutu ile sütunlari birlestirir. en kücük listeyi baz alir.
    print('5:', x, ':', y)                          #   döngü icerisinden birden fazla atama yapilabilir

print('6:', *range(16,0,-2))                        #   verilen degere göre tersten sayi üretebiliriz.

for i in "pythontr":
    if i == "t":
        continue                                    #   burada eger kosul saglaniyorsa alttaki komutlari atlar ve basa döner. While ile de kullanilir
    print(i, end=' ')

cift = []
for i in range(12):
    if i % 2 == 0:
        cift.append(i*i)                            #   cift sayilarin karesini listeye ekleme yaptirdik 
print('\n8:', cift)
yeni = [i*i for i in range(12) if i % 2 == 0]       #   for ve if komutlarini tek satirda kisaltarak yazdik. atama yapamk icin "[]"" yada "liste" formatinda olmali.
print('9:', yeni)                                   #   asil islem basa yazilir sonra tüm satir bosluklar ile yazilir
```

### **WHILE DÖNGÜ KOMUT KULLANIMI**

```python
liste = ['a', 'b', 'c', 'd', 'e']
a = 0
while a < len(liste):                               #   bu komut kosula baglidir. True ve False ile calisir ve döngü kurar. Ilk kosul olarak True mu diye bakar. 
    print('1: {} karesi..: {}'.format(a, a**2))
    a += 1                                          #   her seferinde listeye geri dönmesini ve sayiyi arttirmasini sagladik.

sayi = 45
while True:                                         #   True ifadesi ile sonsuz döngü baslattik
    tahmin = int(input('sayi kac..: '))
    if tahmin < sayi:                               #   döngü icerisinde "if" ifadesi ile yeni kosullar olusturulabilir
        print('2: daha büyük')
    elif tahmin > sayi:
        print('3: daha kücük')
    else:
        print('4: sayiyi buldun')
        break                                       #   döngüden cikmak icin kullanilir

yas = input('Yasinizi girin..:')
while not yas.isdigit():                            #   "isdigit" komutu sadece rakam girilmesine izin verir. String kabul etmez.
    yas = input('Lütfen rakam olarak girin..: ')
print('5: Bravo yasin :', yas)
```

### **RANDOM KOMUTU**

```python
import random
a = random.randint(1,20)                                        #   rastgele bir tamsayi üretmek icin kullanilir. Aralik parantez icinde belirtilir
b = random.uniform(0.5, 2.0)                                    #   rastgele bir kesirli sayi üretmek icin kullanilir. Aralik parantez icinde belirtilir
c = random.random()                                             #   0 - 1 araliginda rastgele bir sayi üretmek icin kullanilir

liste = ['ali', 'veli', 'ahmet', 'mehmet', 'celal']
d = random.choice(liste)                                        #   liste icerisinden rastgele bir kelime secer
e = random.choice(liste[0])                                     #   liste icinde kelimeler arasindan rastgele bir harf secer

f = list(range(11))
random.shuffle(f)                                               #   Listeyi karisik bir sekilde yazdirmak icin kullanilir

s = range(100)
g = random.sample(s, 3)                                         #   belirtilen liste arasindan istenilen sayi kadar rastgele sayi secer

print('1:', a)
print('2:', b)
print('3:', c)
print('4:', d)
print('5:', e)
print('6:', f)
print('7:', g)
```

### **DEF FONKSIYONU KULLANIMI**

```python
b = 5                                     #   burada "b" global bir degiskendir her yerden ulasilabilir ve cagirilabilir
def deger():                              #   "def" komutu ile fonksiyon baslatilir ve fonksiyona bir isim atanir
    global b                              #   burada fonksiyon icindeki "b" degiskeninin her yerden cagirilabilir olmasini sagladik
    b = 7                                 #   fonksiyon icinde deger atamasi yapar
    #print(b)                             #   fonksiyon icindeki "print" sadece fonksiyon icindeki ve yerel degiskenleri yazdirmak icin kullanilir
deger()                                   #   fonksiyonun calistirilmasi disarida fonksiyon isminin yazilmasi ile mümkün
print('1:', b)                            #   fonksiyon disinda oldugu icin sadece global degiskenlerini yazdirir

def sayi():
    s = 5
    return s                              #   "return" komutu ile "sayi()" fonksiyonu bir degere atama yapilabilir ve print() icinde yazdirilabilir
son = sayi()                              #   "return" komutu sayesinde fonksiyon ismi kullanarak sonucu bir degiskene atama yaptik
print('2:', sayi())                       #   "return" fonksiyon ismi ile direk yazdirma yaptik. (Parantezler önemli)
print('3:', son)                          #   fonksiyon cagrilirken kullanilan parametrelere argüman denir

def coklu(sayi, mesaj = 'Hallo', dongu = 3):                #   fonksiyon icerisinde degerler parametredir. Burada otomatik olarak tanimladik.
    print('4:', sayi, 'kez', mesaj, dongu, 'e kadar gider') #   "sayi" parametresi konumsal argümandir ve ilk önce yazilmalidir.
coklu(5)                                                    #   diger degerler girilmese bile tanimli olanlari otomaik olarak yazar
coklu(4, 'Naberr')                                          #   sirasiyla fonksiyon icindeki tanimli degere yeni deger atanabilir
coklu(mesaj='Gece', sayi=8, dongu=10)                       #   sirasiz olarak yeni degerler atanabilir. Birine atanirsa digerine de deger atanmali yoksa hata verir

def meyveler(*fruit):                                       #   " * " kac tane girdi olacagini bilmedigimiz durumlarda kullanilir              
    print('5:', fruit)                                      #   *args  --> keyfi sayida konum argümani tanimlamak icin kullanilir
hazir =['Cilek', 'Karpuz', 'Erik'] 
meyveler('Portakal', 'Elma', 'Armut')                       
meyveler(*hazir)                                            #   Tanimli bir liste icerisinden buraya deger almak icin de " * " kullanilabilir

def hayvan(**animal):                                       #   " ** " kac adet anahtar kelime ve argüman girilecegini bilmedigimiz durumlarda kullanilir
    for i, k in animal.items():                             #   **kwargs  --> keyfi sayida anahtar kelime tanimlama icin kullanilir
        print('6:', k, 'bir', i)  
aile = {'y': 'Murat', 'x': 'Ufuk'}                          
hayvan(etobur='Aslan', otobur='Fil', hepobur='Ayi')         
hayvan(**aile)                                              #   Tanimli bir sözlük icerisinden buraya deger almak icin de " ** " kullanilabilir

number = [1,2,5,8,9,11,12,16,20,25,37,38]
def cift(x):
   return x % 2 == 0                                        #   Çağırılan fonksiyonun döndürdüğü değerin true olduğu durumlara göre liste döndürür
print('7:', list(filter(cift, number)))                     #   liste üzerinde filtreleme yapmamizi saglar. **filter(fonk. adi, liste adi)**

def kare(sayi):
    return sayi * sayi
sayi = range(1,10)                                          #   "map" listenin tüm elemanlarini tek tek "kare" fonksiyonuna gönderir ve geri döndürür.
print('8:', list(map(kare, sayi)))                          #   birden fazla liste alabilir. **map(fonk. adi, liste adi, liste adi)**

def ana():
    x = 'Dis katman'
    def icten():
        nonlocal x                                          #   en icte tanimlanan degiskeni yerel olmaktan cikardi ve kapsamini genisletti
        x = 'Ic katman'
        print('9: Icte: ', x)
    icten()                                                   #   ic ice tanimlanan fonksiyonlarda cikti alabilmek en icten baslayarak kapatmak lazim
    print('9: Dista: ', x)
ana()

print('10: ', bool.__doc__)                                  #   "__doc__" önceden tanimli fonksiyonun görevini ögrenmek icin kullanilir
```
